```
#exposure is extra variable to categorize good and bad loan
lc_df=lc_df.withColumn("exposure",when(lc_df.bad_loan=="No",col("revol_util")).otherwise(-10*col("revol_util")))
display(lc_df)
```

```
#creating udf

from pyspark.sql.types import FloatType

def trim(string):
  return string.strip('%')

spark.udf.register('trimperct',trim)
```

```
%sql

select int_rate,cast(trimperct(int_rate) as float) as int_rate from lc_loan_data;
```

```
%sql

select bad_loan,cast(trimperct(int_rate)as float ) as int_rate_float from lc_loan_data
```
